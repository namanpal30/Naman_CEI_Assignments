```python
import tensorflow as tf
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Embedding, SimpleRNN, LSTM, GRU, Dense
import numpy as np
import matplotlib.pyplot as plt
print("TensorFlow:", tf.__version__)
```

# Load Corpus Text

```python
corpus = '''
machine learning helps computers learn patterns from data without explicit programming
deep learning uses neural networks with many layers to solve complex problems
artificial intelligence is transforming industries by automating tasks and improving decisions
python is one of the most popular programming languages for data science
students can improve their coding skills by practicing algorithms every day
the quick brown fox jumps over the lazy dog near the river
data scientists analyze large datasets to discover meaningful insights and trends
natural language processing enables computers to understand and generate human language
software developers build applications that solve real world business problems efficiently
cloud computing provides scalable resources that can be accessed over the internet
the weather was pleasant and the birds were singing in the morning
a successful project requires planning collaboration testing and continuous improvement
reading books regularly can improve vocabulary critical thinking and communication skills
many companies use machine learning models to predict customer behavior accurately
the university offers courses in programming databases networking and artificial intelligence
learning new technologies can open exciting career opportunities in the future
the football team practiced hard every day before the championship tournament
researchers are developing advanced systems that can assist people in daily life
database management systems store organize and retrieve information efficiently and securely
consistent effort and dedication are important factors for achieving long term success
'''
print(corpus)
```

# Tokenization & Sequence Creation

```python
tokenizer = Tokenizer()
tokenizer.fit_on_texts([corpus])
total_words = len(tokenizer.word_index) + 1
print("Vocabulary Size: ", total_words)
input_sequences = []
for sequences in corpus.split('\n'):
  token_list = tokenizer.texts_to_sequences([sequences])[0]
  for i in range(1, len(token_list)):
    n_gram_seq = token_list[ :i+1]
    input_sequences.append(n_gram_seq)
max_len = max(len(seq) for seq in input_sequences)
input_sequences = pad_sequences(input_sequences, maxlen = max_len, padding='pre')
X = input_sequences[:,:-1]
y = input_sequences[:, -1]
print("X shape:", X.shape)
print("y shape:", y.shape)
```

# Vanilla RNN


```python
rnn_model = Sequential([
    Embedding(total_words, 32, input_length=max_len-1),
    SimpleRNN(64),
    Dense(total_words, activation='softmax')
])

rnn_model.compile(loss='sparse_categorical_crossentropy',
                  optimizer='adam',
                  metrics=['accuracy'])

rnn_history = rnn_model.fit(X, y, epochs=100, verbose=0)
print("Vanilla RNN training completed")
```

# LSTM

```python
lstm_model = Sequential([
    Embedding(total_words, 32, input_length=max_len-1),
    LSTM(64),
    Dense(total_words, activation='softmax')
])

lstm_model.compile(loss='sparse_categorical_crossentropy',
                   optimizer='adam',
                   metrics=['accuracy'])

lstm_history = lstm_model.fit(X, y, epochs=100, verbose=0)
print("LSTM training completed")
```

# GRU

```python
gru_model = Sequential([
    Embedding(total_words, 32, input_length=max_len-1),
    GRU(64),
    Dense(total_words, activation='softmax')
])

gru_model.compile(loss='sparse_categorical_crossentropy',
                  optimizer='adam',
                  metrics=['accuracy'])

gru_history = gru_model.fit(X, y, epochs=100, verbose=0)
print("GRU training completed")
```

# Compare Training Loss
* 100 Epochs

```python
plt.figure(figsize=(10,4))
plt.plot(rnn_history.history['loss'], label='RNN')
plt.plot(lstm_history.history['loss'], label='LSTM')
plt.plot(gru_history.history['loss'], label='GRU')
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.title("Training Loss Comparison")
plt.legend()
plt.show()
```

# Text Generation Function

```python
def generate_text(model, seed_text, next_words=5):
    for _ in range(next_words):
        token_list = tokenizer.texts_to_sequences([seed_text])[0]
        token_list = pad_sequences([token_list], maxlen=max_len-1, padding='pre')
        predicted = np.argmax(model.predict(token_list, verbose=0), axis=-1)[0]

        output_word = ""
        for word, index in tokenizer.word_index.items():
            if index == predicted:
                output_word = word
                break
        seed_text += " " + output_word
    return seed_text
```

# Sample Text
* words - 5

```python
print("RNN :", generate_text(rnn_model, "deep learning", 5))
print("LSTM:", generate_text(lstm_model, "deep learning", 5))
print("GRU :", generate_text(gru_model, "deep learning", 5))
```
